```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
a= np.linspace(0,5,10)
b=a**3
```

```python
a
```

```python
b
```

```python
plt.plot(a,b)
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r')
plt.subplot(1,2,2)
plt.plot(b,a,'b')
```

```python
figure=plt.figure()
axes=figure.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
axes.set_xlabel('a array value')
axes.set_ylabel('b array value')
axes.set_title('Some title')
axes.plot(a,b)
plt.figure()
```
